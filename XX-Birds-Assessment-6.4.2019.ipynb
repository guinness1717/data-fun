```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sn
```

```python
file_path = '~/Desktop/ES-Misc/Data-Python/XX-Birds/form-6.4.2019.csv'
```

```python
df = pd.read_csv(file_path, ',')
```

```python
df.head()
```

```python
df_drop = df.drop(['Finish time','Layers','Layer thickness (um)','Z offset (mm)','Open mode?','Notes'], axis=1)
```

```python
df_rename = df_drop.rename(columns={'Print name':'Name',
                                   'Start time':'Date',
                                   'Elapsed print time (ms)':'Duration',
                                   'Volume (ml)':'Volume'})
```

```python
df_TEC = df_rename.loc[df_rename.Printer.isin(['EvanescentFish','FanaticalMule'])]
```

```python
df_TEC.Date = pd.to_datetime(df_TEC.Date, format='%Y%m%d %H:%M:%S')
```

```python
df_TEC.Date = df_TEC.Date.dt.date
```

```python
df_TEC.Date = pd.to_datetime(df_TEC.Date, format='%Y-%m-%d')
```

```python
df_index = df_TEC.set_index('Date')
```

```python
df_sort = df_index.loc['2019-06-03':'2019-05-01']
```

```python
df_sort.tail()
```

```python
df_XX = df_sort.loc[df_sort.Material.isin(['Rigid V1'])]
```

```python
df_XX.tail(10)
```

```python
df_finished = df_XX[df_XX.Status != 'ABORTED']
```

```python
df_finished.head()
```

```python
df_prep = df_finished.reset_index()
```

```python
df_prep.Duration = df_prep.Duration.apply(lambda x: x/3600000)
```

```python
df_round = df_prep.round({'Duration': 4})
```

```python
pd.pivot_table(df_round, values=['Duration','Volume'], index='Printer', aggfunc={'Duration': np.sum,
                                                                             'Volume': np.sum}, margins=True,
                                                                              margins_name='Total')
```
